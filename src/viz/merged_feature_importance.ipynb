```python
import pandas as pd
import numpy as np
```

```python
feat1 = pd.read_table("./feature_importance_permutation_test1.tsv",sep="\t")
feat2 = pd.read_table("./feature_importance_permutation_test2.tsv",sep="\t")
feat3 = pd.read_table("./feature_importance_permutation_test3.tsv",sep="\t")
feat1.head()
```

```python
feat1['rank'] = feat1['importance'].rank(ascending=False)
feat2['rank'] = feat2['importance'].rank(ascending=False)
feat3['rank'] = feat3['importance'].rank(ascending=False)
```

```python
feat3.head()
```

```python
feat12 = feat1.set_index("GO").join(feat2.set_index('GO'), on='GO', lsuffix='_1', rsuffix='_2')
feat12
```

```python
feat123 = feat12.join(feat3.set_index('GO'), on='GO')
feat123
```

```python
feat123['mean_rank'] = feat123[['rank_1', 'rank_2', 'rank']].mean(axis=1)
feat123['mean_rank']
```

```python
feat123 = feat123.sort_values(by='mean_rank', ascending=True)
feat123
```
